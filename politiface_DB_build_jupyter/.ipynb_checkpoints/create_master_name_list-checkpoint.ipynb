```python
# https://theunitedstates.io/congress-legislators/legislators-current.json has a big json table of reps
```

```python
json = pd.read_json("https://theunitedstates.io/congress-legislators/legislators-current.json")
```

```python
curr_congress_raw = pd.json_normalize(json['name'])
```

```python
curr_congress_raw['export_name_1'] = ""
curr_congress_raw['export_name_2'] = ""
curr_congress_raw = curr_congress_raw.drop("official_full", axis=1)
```

```python
def strip_punct(x):
    if type(x[2]) == str:
        x[2] = x[2].replace(".","")
    return x



def create_official(x):
    if (type(x[2]) == float) & (type(x[3]) == float):
        x[6] = (x[0]+" "+x[1])
    if (type(x[2]) == float):
        x[6] = (x[0]+" "+x[3]+" "+x[1])
    return x
    

curr_congress_raw = curr_congress_raw.apply(lambda x: strip_punct(x), axis=1)
```
